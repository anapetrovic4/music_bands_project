## Computation of measures of shape and distribution

### Skewness

```python
df['duration'].skew()
```

Since the skewness is greater than 1, it can be concluded that the duration data is highly skewed.

### Kurtosis

```python
kurt_value = df['duration'].kurt()
kurt_value
```

Since the kurtosis is greater than 3, it can be concluded that the duration curve is a leptokurtic curve. As compared to a normal distribution, its central peak is higher and sharper, and its tails are longer and fatter.

```python
sample_data = df['duration'].sample(n=774, random_state=1)

# Prikaz distribucije uzorka podataka
sns.histplot(sample_data, kde=True)
plt.title(f'Kurtosis: {sample_data.kurt():.2f}')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.show()
```
