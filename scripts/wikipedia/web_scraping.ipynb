## Crosby, Stills, Nash | Crosby, Stills, Nash & Young

```python
# Function to extract all album links from a webpage
def get_albums_links(url):
  data = requests.get(url) # Send a GET request to the specified URL
  soup = BeautifulSoup(data.content, 'html.parser') # Parse the content using BeautifulSoup
  ul_elements = soup.find_all('ul')
  songs = []

  # Iterate through a list of <ul> tags
  for i, ul in enumerate(ul_elements):
    li_elements = ul.find_all('li') # Find all <li> tags within <ul> list
    if(i == 40): # ***Note***: Because the list with discography information isn't defined via class or id, it is located based on the index position.
      for li in li_elements:
        a_elements = li.find_all('a') # Find all anchor tags that contain links to the albums
        for a in a_elements:
          href = a.get('href')
          songs.append(f'https://en.wikipedia.org{href}'); # Store the results in a list 'songs'

  return songs # Return the list of songs
```

### Crosby & Nash

```python
# Function to extract all album links for Crosby & Nash
def get_all_albums(url):
  data = requests.get(url) # Send a GET request to the specified URL
  soup = BeautifulSoup(data.content, 'html.parser') # Parse the content using BeautifulSoup
  tables = soup.find_all('table', class_='wikitable') # Find all tables with class 'wikitable'
  songs = soup.find_all('td', {'scope': 'row'}) # Find all <td> elements where 'scope' is set to 'row'
  song_links = ['https://en.wikipedia.org/' + song.find('a')['href'] for song in songs if song.find('a')] # Create a list of song links by extracting the 'href' attribute from the <a> tags

  return song_links # Return song links
```

## Function for extracting information

```python
# Function to extract additional information related to tracks and albums
def extract_information(url):

  release_date = ''
  recording_period = ''
  producers = ''
  writers = ''
  track_name = ''
  album = ''
  lead_vocals = ''
  track_names = []
  writers = []
  lead_vocals = []

  response = requests.get(url) # Send a GET request to the specified URL
  soup = BeautifulSoup(response.content, 'html.parser') # Parse the content using BeautifulSoup

  table = soup.find('table', class_='infobox vevent haudio') # Find table with basic information, such as recording period, release date and producer names
  if table:
    album_th = table.find('th', class_='infobox-above summary album') # Extract album name
    if album_th:
      album = album_th.get_text(strip=True).strip('"')

    # Function to extract text from given title of table header
    def extract_text(th_name):
      th = table.find('th', string=th_name)
      return th.find_next_sibling('td').get_text(strip=True) if th else None

    release_date = extract_text('Released')
    recording_period = extract_text('Recorded')

    # Store producers in a list. If there is more than one producer, join them with comma
    producer_row = table.find('th', string='Producer')
    if producer_row:
      producer_column = producer_row.find_next_sibling('td')
      producer_list = []

      a_elements = producer_column.find_all('a')
      if a_elements:
          for a in a_elements:
            producer_list.append(a.get_text(strip=True))

      if not producer_list:
          producer_text = producer_column.get_text(strip=True)
          producer_list = [producer.strip() for producer in producer_text.split(',') if producer.strip()]

      producers = ', '.join(producer_list)

  # Find all tables with class 'tracklist', which contain data about track name, composers, writers and lead vocals
  track_listing = soup.find_all('table', class_='tracklist')

  if track_listing:
    for track_list in track_listing:

      rows = track_list.find_all('tr')

      # Create lists of booleans to check whether lead vocals or writers exist in given table.
      headers = [th.get_text(strip=True).lower() for th in track_list.find_all('th')]
      lead_vocals_exists = ['lead vocals' in header for header in headers]
      writers_exist = ['writer' in header for header in headers]

      for row in rows[1:-1]: # ignore header and footer TODO do not ignore header and footer

        columns = row.find_all('td')

        track_name = columns[0].get_text(strip=True).replace('"', '')
        writer_string = None
        lead_vocal_string = None

        for writer_exist in writers_exist:
          if writer_exist == True:
            writer_column = columns[1]
            writer_list = []

            li_elements = writer_column.find_all('li')
            if li_elements:
              for a in li_elements:
                writer_list.append(a.get_text(strip=True))

            if not li_elements:
              writer_text = writer_column.get_text(strip=True)
              writer_list = [writer.strip() for writer in writer_text.split(',') if writer.strip()]

            writer_string = ', '.join(writer_list)
            break

        for lead_vocal in lead_vocals_exists:
          if lead_vocal == True:
            if len(columns) > 2:
              lead_vocal_string = columns[2].get_text(strip=True)
            break

        track_names.append(track_name)
        writers.append(writer_string)
        lead_vocals.append(lead_vocal_string)


  return {
    'track' : track_names,
    'album' : album,
    'release_date' : release_date,
    'recording_period' : recording_period,
    'producers' : producers,
    'writers' : writers,
    'vocals' : lead_vocals,
  }
```

```python
# Function to collect data for each artist
def collect_data(songs, filename, list):
  for index, song in enumerate(songs):
    if ')' in song:
      song = song.replace("',", "") # *** Note ***: In case of Neil Young studio songs, two last values are treated as parts of external URLs, so they are removed to ensure correct data. TODO
      song_data = extract_information(song)

      df = pd.DataFrame(song_data)
      list.append(df)

    else:
      song_data = extract_information(song)

      df = pd.DataFrame(song_data)
      list.append(df)

  all_songs_df = pd.concat(list, ignore_index=True) # Concat multiple dataframes to final dataframe

  all_songs_df.to_csv(f'{filename}.csv', index=False) # Convert dataframe to .csv file
  return all_songs_df
```

## Data cleaning

```python
# Format dates: remove date format between round brackets ()
pattern_date = r'\([^)]*\)'
df_wikipedia['release_date'] = df_wikipedia['release_date'].str.replace(pattern_date, '', regex=True)
df_wikipedia['recording_period'] = df_wikipedia['recording_period'].str.replace(pattern_date, '', regex=True)

# Remove recording locations which are placed in recording period
pattern_locations = r'^[^0-9]*$'
df_wikipedia['recording_period'] = df_wikipedia['recording_period'].str.replace(pattern_locations, '', regex=True)

# Remove square brackets []
pattern_brackets = r'\[\d+\]'

df_wikipedia['release_date'] = df_wikipedia['release_date'].str.replace(pattern_brackets, '', regex=True)
df_wikipedia['recording_period'] = df_wikipedia['recording_period'].str.replace(pattern_brackets, '', regex=True)
df_wikipedia['track'] = df_wikipedia['track'].str.replace(pattern_brackets, '', regex=True)
df_wikipedia['producers'] = df_wikipedia['producers'].str.replace(pattern_brackets, '', regex=True)

# Remove numbers from lead_vocals
pattern_numbers = r'^\d:\d{2}$'
df_wikipedia['vocals'] = df_wikipedia['vocals'].str.replace(pattern_numbers, '', regex=True)

# Format writers: remove words traditional or arrangement and keep only names
pattern_writers = r'\b(traditional|arrangement)\b'
df_wikipedia['writers'] = df_wikipedia['writers'].apply(
    lambda text: re.sub(pattern_writers, '', text, flags=re.IGNORECASE).strip() if isinstance(text, str) else text
)

df_wikipedia = df_wikipedia.fillna('')

df_wikipedia.to_csv('wikipedia.csv', index=False)
df_wikipedia
```
