## Merge all DataFrames to 'allmusic-v3.csv'

```python
allmusic_v4 = pd.concat([david_crosby_allmusic,
                         stephen_stills_allmusic,
                         graham_nash_allmusic,
                         neil_young_allmusic,
                         stills_young_allmusic,
                         crosby_nash_allmusic,
                         crosby_stills_nash_allmusic,
                         crosby_stills_nash_young_allmusic
                         ]).reset_index(drop=True)

allmusic_v4['track_lower'] = allmusic_v4['track'].str.lower() # create additional column with track titles in lower case
allmusic_v4.sort_values(by=['track_lower', 'album_type'], ascending=[True, False], inplace=True)
allmusic_v4 = allmusic_v4.drop_duplicates(subset=['track_lower'], keep='first') # delete duplicates based on 'track_lower'
allmusic_v4.drop(columns=['track_lower'], inplace=True) # delete no longer needed column

allmusic_v4 = allmusic_v4[~allmusic_v4['track'].str.startswith('Looking for a Leader')] # delete duplicated track

allmusic_v4.to_csv('allmusic-v4.csv', index=False)
allmusic_v4.reset_index(drop=True, inplace=True)
allmusic_v4
```
