  ## David Crosby

```python
def clean_dataset(path):
  df_allmusic = pd.read_csv(path)

  # Rename the columns
  df_allmusic.rename(columns={'track_name' : 'track', 'album_name' : 'album', 'album_release_date' : 'release_date', 'album_recording_location' : 'recording_location', 'album_recording_period' : 'recording_period'}, inplace=True)

  # Change order of the columns
  new_order = ['track', 'album', 'album_type', 'artist', 'album_duration', 'release_date', 'recording_period', 'recording_location', 'album_genre', 'album_styles', 'album_moods',	'album_themes', 'genres', 'styles', 'moods', 'themes', 'composers']
  df_allmusic = df_allmusic[new_order]

 # Format release_date to yyyy-dd-mm
  def convert_date_format(date_string):
      if date_string is None or pd.isna(date_string) or date_string == '':
        return '' # If the value is None, NaN ili empty string, return empty string
      try:
          return pd.to_datetime(date_string, format='%B %d, %Y').strftime('%Y-%d-%m') # Try to convert format "Month Day, YYYY"
      except ValueError:
          try:
              return pd.to_datetime(date_string, format='%B, %Y').strftime('%Y-%m') # Try to convert format "Month, YYYY"
          except ValueError:
              try:
                  return pd.to_datetime(date_string, format='%Y').strftime('%Y') # Try to convert format "YYYY"
              except ValueError:
                  return ''

  df_allmusic['release_date'] = df_allmusic['release_date'].apply(convert_date_format)

  df_allmusic.sort_values(by=['track', 'album_type'], ascending=[True, False], inplace=True)

  # Delete duplicates based on 'track'
  df_allmusic['track'] = df_allmusic['track'].replace('Déjà Vu', 'Deja Vu') # rename for easier access
  df_allmusic['track'] = df_allmusic['track'].replace('What Are Their Names', 'What Are Their Names?') # rename for easier access


  df_allmusic = df_allmusic.drop_duplicates(subset=['track'], keep='first')
  df_allmusic.reset_index(drop=True, inplace=True)

  return df_allmusic

david_crosby_allmusic = clean_dataset('/content/allmusic-david-crosby-v2.csv')
david_crosby_allmusic.to_csv('allmusic-david-crosby-v3.csv', index=False)
david_crosby_allmusic
```

## Stephen Stills

```python
stephen_stills_allmusic = clean_dataset('/content/allmusic-stephen-stills-v2.csv')

stephen_stills_allmusic['track_lower'] = stephen_stills_allmusic['track'].str.lower() # create additional column with track titles in lower case
stephen_stills_allmusic.sort_values(by=['track_lower', 'album_type'], ascending=[True, False], inplace=True)
stephen_stills_allmusic = stephen_stills_allmusic.drop_duplicates(subset=['track_lower'], keep='first') # delete duplicates based on 'track_lower'
stephen_stills_allmusic.drop(columns=['track_lower'], inplace=True) # delete no longer needed column

# Delete duplicated rows manually
stephen_stills_allmusic = stephen_stills_allmusic[stephen_stills_allmusic['track'] != '4+20']
stephen_stills_allmusic = stephen_stills_allmusic[stephen_stills_allmusic['track'] != "49 Bye Byes/For What It's Worth"]
stephen_stills_allmusic = stephen_stills_allmusic[stephen_stills_allmusic['track'] != "Blind Fiddler"]
stephen_stills_allmusic = stephen_stills_allmusic[stephen_stills_allmusic['track'] != "Everybody's Talking"]
stephen_stills_allmusic = stephen_stills_allmusic[stephen_stills_allmusic['track'] != "Wounded World/Rocky Mountain Way"]

stephen_stills_allmusic['track'] = stephen_stills_allmusic['track'].replace('Ole Man Trouble', 'Old Man Trouble') # fix spelling mistakes

stephen_stills_allmusic.reset_index(drop=True, inplace=True)
stephen_stills_allmusic.to_csv('allmusic-stephen-stills-v3.csv', index=False)
stephen_stills_allmusic
```
