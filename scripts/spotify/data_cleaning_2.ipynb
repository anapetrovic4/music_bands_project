```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

## Load DataFrames

```python
spotify_david_crosby_df = pd.read_csv('/spotify-david-crosby-v3.csv')
spotify_graham_nash_df = pd.read_csv('/spotify-graham-nash-v3.csv')
spotify_neil_young_df = pd.read_csv('/spotify-neil-young-v3.csv')
spotify_stephen_stills_df = pd.read_csv('/spotify-stephen-stills-v3.csv')
spotify_stills_young_df = pd.read_csv('/spotify-stills-young-band-v3.csv')
spotify_cn_df = pd.read_csv('/spotify-cn-v3.csv')
spotify_csn_df = pd.read_csv('/spotify-csn-v3.csv')
spotify_csny_df = pd.read_csv('/spotify-csny-v3.csv')
```

## Concatenate loaded DataFrames into 'spotify-v3.csv'

```python
spotify_v3_df = pd.concat([spotify_david_crosby_df, spotify_graham_nash_df, spotify_neil_young_df,
                           spotify_stephen_stills_df, spotify_stills_young_df, spotify_cn_df,
                           spotify_csn_df, spotify_csny_df])

spotify_v3_df.reset_index(drop=True, inplace=True)
spotify_v3_df
```

## Clean duplicated songs

```python
spotify_v3_df['track_lower'] = spotify_v3_df['track'].str.lower() # Converting all letters to lower case to ensure correct comparison
spotify_v3_df.sort_values(by=['track_lower', 'album_type'], ascending=[True, False], inplace=True) # Sort the DataFrame by 'track' and 'album_type' to ensure keeping only the studio version
spotify_v3_df = spotify_v3_df.drop_duplicates(subset=['track_lower'], keep='first') # Keep the track where 'album type' is 'studio'
spotify_v3_df.drop(columns=['track_lower'], inplace=True) # Remove column 'track_lower', as it is no longer needed
spotify_v3_df.reset_index(drop=True, inplace=True)
spotify_v3_df.to_csv('/spotify-v3.csv', index=False)
spotify_v3_df
```
